```python
pook= pd.read_excel("Pokemon.xlsx")
ates=pook['Type1']=='Fire'
pook[ates]
 
 #İstenilen türe uyan pokemonları bir değişkene atayabiliriz.Bu değişkeni daha sonra kullanabiliriz.
```

```python
drg=pook['Type2']=='Dragon'
pook[ates | drg ]

#iki farklı koşulu sağlayanları alır
# | işareti veya demektir.
```

```python
pook[ates & drg ]
#& işareti ve demektir.
```

```python
hiz120=pook['Speed']>120
pook[hiz120]
#hızı 120 den büyük olan pokemonları alırız.
```
